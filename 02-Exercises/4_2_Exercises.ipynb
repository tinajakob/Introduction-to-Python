### Exercise 1.1

Run the following code to fetch the Wikipedia text about dogs:

```
import requests
from bs4 import BeautifulSoup as bs
import re

# Get page and convert to json
r = requests.get("https://en.wikipedia.org/w/api.php?action=parse&page=Dog&format=json").json()

# Get text and convert to BeautifulSoup object
r_text = r["parse"]["text"]["*"]
soup = bs(r_text)

# Get list of all paragraphs 
pars = soup.find_all("p")

# Remove HTML tags (get text)
pars = [par.get_text() for par in pars]
        
# Join paragraphs to one string        
dog = ' '.join(pars)

# Remove citations and punctuation
dog = re.sub(r"\[\d+\]", "", dog) # remove citations
dog = re.sub(r"[^\w\s-]", "", dog) # remove all characters that are not word characters, spaces or -
dog
```
